```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def bazaraa_shetty(x,y):
    return (x-2)**4 + (x-2*y)**2

def roosenbrock(x,y):
    return 100*(y-(x)**2)**2 + (1-x)**2

x = np.linspace (-2.    # lower limit
                , 3     # upper limit
                , 21  # generate 51 points between 0 and 3
                )
y = np.linspace (-2.    # lower limit
                , 3     # upper limit
                , 21    # generate 51 points between 0 and 3
                )

def gradient_bazaraa_shetty(x,y):
    derivative_x = 4*(x-2)**3 + 2*(x-(2*y))
    derivative_y = 2*(x-(2*y))*(-2)
    gradient = np.array([derivative_x,derivative_y], dtype=float)
    return gradient
    
def gradient_roosenbrock(x,y):
    derivative_x = 200*(-2*x)*(y-x**2)+(2*x)-2
    derivative_y = 200*(y-x**2)
    gradient =  np.array([derivative_x,derivative_y], dtype=float)
    return gradient
```

```python
def check_condition1(old_input,new_input,step_size,delta):
    new_function_value = float(roosenbrock(new_input[0],new_input[1]))
    old_function_value = float(roosenbrock(old_input[0],old_input[1]))
    check = new_function_value <= old_function_value + (-1*step_size*np.linalg.norm(gradient_roosenbrock(old_input[0],old_input[1])))
    return check

def check_condition2(old_input,new_input,step_size,delta):
    new_function_value = float(roosenbrock(new_input[0],new_input[1]))
    old_function_value = float(roosenbrock(old_input[0],old_input[1]))
    check = new_function_value >= old_function_value + (-1*step_size*np.linalg.norm(gradient_roosenbrock(old_input[0],old_input[1])))
    return check

def armijoLineSearch(step_size,delta,func,initial_input):
    old_input = initial_input
    new_input = np.add(old_input, np.array([-1*step_size,-1*step_size],dtype=float))
    
    is_condition1_true = check_condition1(old_input,new_input,step_size,delta)
    is_condition2_true = check_condition2(old_input,new_input,step_size,delta)
    current_step_size = step_size
    print(is_condition1_true)
    print(is_condition2_true)
    
    #print('step after condition1',current_step_size)
    
    while (is_condition1_true):
        current_step_size = current_step_size*2
        old_input = new_input
        new_input = np.add(old_input, np.array([-1*current_step_size,-1*current_step_size],dtype=float))
        is_condition1_true = check_condition1(old_input,new_input,current_step_size,delta)
#         print(is_condition1_true)
    
    while not(is_condition1_true):
        current_step_size = current_step_size/2
        old_input = new_input
        new_input = np.add(old_input, np.array([-1*current_step_size,-1*current_step_size],dtype=float))
        is_condition1_true = check_condition1(old_input,new_input,current_step_size,delta)
#         print(current_step_size)
      
    #print('step after condition2',current_step_size)
    return current_step_size

i = armijoLineSearch(1,.0001,"roosenbrock",np.array([-200.00,400.15],dtype=float))
print(i)
```

```python
def decent_method(initial_step_size,delta,func,coordinate):
    is_stopping_creiterion_true = np.linalg.norm(np.gradient(coordinate))==0
    print(is_stopping_creiterion_true)
    while not(is_stopping_creiterion_true):
        step_size = armijoLineSearch(initial_step_size,delta,func,coordinate)
        new_input = np.add(coordinate, np.array([-1*step_size,-1*step_size],dtype=float)) 
        coordinate = new_input
        initial_step_size = step_size
        is_stopping_creiterion_true = np.linalg.norm(np.gradient(coordinate))==0
        print(step_size)
    #return (roosenbrock(coordinate))
    
```
