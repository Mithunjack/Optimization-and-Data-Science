```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import fft
from matplotlib.image import imread
from matplotlib.pyplot import imshow

A = imread('easter.PNG')
gs_image = np.mean(A, -1)

# np.shape(A)
# imshow(A)
# plt.axis('off')
```

```python
coef_array = fft.fft2(A)
sorted_coef_array = np.argsort(coef_array)[::-1]
for keep in (1,0.1,0.05,0.02,0.01,0.0001):
    i = 1
    thresh = sorted_coef_array[int(np.floor((1-keep)*len(sorted_coef_array)))]
    print(len(sorted_coef_array))
    
    ind = np.abs(coef_array)>thresh
    atlow = coef_array*ind
    co = fft.ifft2(atlow).real
   
    plt.figure(i)
    plt.imshow(co)
    plt.legend()
    i += 1

    
```
