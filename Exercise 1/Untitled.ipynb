```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import fft

time_step = 0.05
time_vec = np.arange(0,10,time_step)

period = 5

sig = np.round((np.sin(2*np.pi*time_vec/period)),2)
nois_sig = np.round((np.sin(2*np.pi*time_vec/period)+0.25*np.random.rand(time_vec.size)),2)

sig_fft = fft.fft(nois_sig)

amplititude = np.abs(sig_fft)

sample_freq = fft.fftfreq(nois_sig.size, d=time_step)
amp_freq = np.array([amplititude,sample_freq])

max_amp_position = amp_freq[0,:].argmax()
print(amp_freq[0,:])
peak_freq = amp_freq[1,max_amp_position]



print(peak_freq)
print(max_amp_position)

high_freq_fft = sig_fft.copy()
high_freq_fft[np.abs(sample_freq) > peak_freq]  = 0
filtered_sig = fft.ifft(high_freq_fft)

# plt.plot(nois_sig)
# plt.plot(filtered_sig)
plt.plot(sig)

```
