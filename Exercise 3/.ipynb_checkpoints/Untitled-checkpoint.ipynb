```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def bazaraa_shetty(x,y):
    return (x-2)**4 + (x-2*y)**2

def roosenbrock(x,y):
    return 100*(y-(x)**2)**2 + (1-x)**2
x = np.linspace (-5.    # lower limit
                , 5.     # upper limit
                , 1001  # generate 51 points between 0 and 3
                )
y = np.linspace (-5.    # lower limit
                , 5     # upper limit
                , 1001    # generate 51 points between 0 and 3
                )
func1 = roosenbrock(x,y)
# print(func1)
# print(x)
func2 = bazaraa_shetty(x,y)
plt.plot(x, func1)
plt.show()
plt.plot(x, func2)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection='3d')

# Make data.
X, Y = np.meshgrid(x, y)
func1 = bazaraa_shetty(X, Y)

# Plot the surface.
surf = ax.plot_wireframe(X, Y, func1)

plt.show()

fig = plt.figure(figsize = (10,10))
ax = fig.gca(projection='3d')


func2 = roosenbrock(X, Y)

# Plot the surface.
surf = ax.plot_wireframe(X, Y, func2)

plt.show()
```
