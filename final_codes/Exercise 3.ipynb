### Visualization of Roosenbrock function

the following function is called the Roosenbrock function

f :$R^2$ →***R***,$f(x, y) := 100(y − x^2)^2+ (1 − x)^2$

we will viusalize it and mark the global minimum at $(x,y)=(1,1)$

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def draw_function(x,y,f,extrema_coord):
    fig = plt.figure(figsize=(20, 10))
    ax = plt.gca(projection='3d')
    s=0.5
    x,y = np.meshgrid(x,y)
    z= f(x,y)
    ax.plot_surface(x, y, z)
    ax.scatter(extrema_coord[0],extrema_coord[1],extrema_coord[2],c='red',s=1000)
```

```python
roosenbork = lambda x,y: (1-x)**2 + 100*(y-x**2)**2
minima_roosenbork = (1,1,0)
x= np.arange(-2, 2.+s, s)
y=np.arange(-2, 2.+s, s)
draw_function(x,y,roosenbork,minima_roosenbork)
```
