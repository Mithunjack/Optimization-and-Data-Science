```python
def gradient(gradient_f,values):
    grad = []
    for function in gradient_f:
        gradient = function(*values)
        grad.append(gradient)
#     print('gradient--->',grad)
    return np.array(grad)
```

```python
def SGD(data_set,inital_value,step,gradient_loss_f,input_function,max_iteration):
    xk = inital_value
    for i in range(0,max_iteration):
        random_tuple = random.choice(data_set)
        
        f_value = input_function(random_tuple[0],xk)
        dk = -gradient_loss_f(f_value,random_tuple[1])
        
        step_size = step(i) if callable(step) else step
        
        xk1 = xk+step_size*dk
        xk  = xk1
        
```

```python
def SGD_with_minibatch(data_set,inital_value,step,gradient_loss_f,input_function,max_iteration):
    xk = inital_value
    xk1 = 0
    minibatch_size = len(data_set)/3
    for i in range(0,max_iteration):
        random_tuple = np.random.choice(data_set, minibatch_size)
        
        vectorize_f = np.vectorize(input_function)
        vectorized_gradient_loss_f = np.vectorize(gradient_loss_f)
        
        f_value = input_function(random_tuple[:,0],xk)
        
        gradient  = vectorized_gradient_loss_f(f_value,random_tuple[:,1])
        dk= -np.mean(gradient)
        
        step_size = step(i) if callable(step) else step
        
        xk1 = xk+step_size*dk
        xk  = xk1
```

```python
def select_next_generation(population,objective_fn,size):
    objective_fn = np.vectorize(objective_fn) #vectorizing objective function for the puprpose of using broadcasting over the population
    population_score = objective_fn(population) #output for each of the population sample
    population_with_score = list(zip(population,population_score)) # combining the population and their score in [(population,score)]
    print('population with score', population_with_score)
    population_with_score.sort(key=lambda x: x[:-1]) #sorting the  
    population_sorted_by_score = np.asarray(population_with_score) #convert array of tuples to ndarray
    fitest = population_sorted_by_score[:size,:-1]
    return fitest
```

```python
def meu_lambda_es_with_msc(number_of_parents,number_of_offsprings,objective_function,stopping_generation_size, max_generation):
    number_of_params_of_ofn = len( getfullargspec(objective_function).args)
    generation_unchanged_counter = 0
    number_of_iteration = 0;
    generation_record = {}
    meu = number_of_parents
    lemda = number_of_offsprings
    mutability = 0.1
    population = np.random.uniform(low= 1, high= meu, size= (meu,number_of_params_of_ofn)) #generate initial population
    while True: #this loop has to be replaced with a loop with termination condition where it would end if the result is not chaning for a specific numnber of generation
        generation_record[number_of_iteration] = population
        indices = np.random.choice(population.shape[0], lemda)

        parents = population[indices] #choosing random number of parents
        z = np.random.normal(loc=0,scale=1,size=(lemda,number_of_params_of_ofn)) #creating random vector for mutation
        offspring = (parents +(mutability * z)) # mutation and ofspirng generation
        population = np.append(population, offspring) #new generation containing meu+lemda population

        population = select_next_generation(population,objective_function,meu) #survival of the fittest
        
        # termination conditions
        generation_comparison = generation_record[number_of_iteration] == population
        if generation_comparison.all():
            generation_unchanged_counter +=1
        else:
            generation_unchanged_counter = 0

        if generation_unchanged_counter == stopping_generation_size:
            break
        elif number_of_iteration == max_generation:
            break
        
        number_of_iteration += 1
    
   
```
